# REVENUE MANAGEMENT

```python
df.sample(10)
```

```python
# ordering the month from January to December

df2 = df.copy()
month_number = {
    "January": 1, "February": 2, "March": 3, "April": 4, "May": 5, "June": 6,
    "July": 7, "August": 8, "September": 9, "October": 10, "November": 11, "December": 12
}

df2["month_number"] = df2["arrival_date_month"].map(month_number)

df2["Date"] = df["arrival_date_year"].astype(str) + "-" +df["arrival_date_month"] + "-" + df["arrival_date_day_of_month"].astype(str)
df2["Date"] = pd.to_datetime(df2["Date"], errors="coerce")
df2 = df2.sort_values("month_number")
df2
```

```python
# trend of adr over month 

fig, ax = plt.subplots(figsize = (11,5))
sns.lineplot(x = df2["arrival_date_month"], y = df2["adr"])
plt.show()
# trend of adr over month 
```

```python
# trend of adr over date 

fig, ax = plt.subplots(figsize = (11,5))
sns.lineplot(x = df2["arrival_date_day_of_month"], y = df2["adr"], marker = 'o')
plt.show()
```

```python
# trend of adr over a week

fig, ax = plt.subplots(figsize = (11,5))
sns.boxplot(x = df2["Date"].dt.day_name(), y = df2["adr"], order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.show()
```

```python
# INTERPRETATION -- July and August has the highest adr and we have low adr on January 
                  # Sunday and monday has the lowest adr when compared to Wednesday (business trip, meeting) 
                  # and Saturday (Weekend - family, tour)
```

```python
# 2
```

```python
df["revenue"] = df["adr"] * (df["stays_in_weekend_nights"] + df["stays_in_week_nights"])
df
```

```python
sns.regplot(x = df["adr"], y = df["revenue"])
plt.title("adr vs total revenue")
plt.show()
```

```python
# adr has positive relationship with revenue when adr increases the revenue also increases. 
# Pricing Strategy -- Many bookings happen on July and August so rooms are in demand on that period so we can increase 
                   #  amount on rooms by 1 or 2 % so that it wont affect booking.
                   # Less bookings happen on January and November so that we can give offer on extra stay or breakfast or lunch to increase over revenue. 
```
